```python
def get_total_images(emotion: str):
    df_emotion = pd.read_csv(f"dataset/{emotion}_filtered.csv")

    total = 0
    # iterate over each row of df
    for index, row in df_emotion.iterrows():
        subject_id = row["subject_id"]
        session_id = row["session_id"]

        # find the corresponding row in df_ckplus with the subject id and session id
        ckplus_row = df_ckplus.loc[
            (df_ckplus["subject_id"] == subject_id)
            & (df_ckplus["session_id"] == session_id)
        ]

        # print(subject_id, session_id, len(ckplus_row))

        total = total + len(ckplus_row)
        
    return total
```

```python
emotions = ["anger", "disgust", "fear", "happiness", "sadness", "surprise"]
```

```python
for emotion in emotions:
    print(emotion, get_total_images(emotion))
```
