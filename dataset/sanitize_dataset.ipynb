```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/disgust_all.csv')
```

```python
df
```

```python
# get all unique subject_id, session_id, ckplus_labelled tuples and only keep these fields
df1 = df[['subject_id', 'session_id', 'ckplus_labelled']].drop_duplicates()
```

```python
df1
```

```python
counts = df.groupby(["subject_id", "session_id"]).size().reset_index(name='count')['count'].values.tolist()
```

```python
# append counts as a column to df1
df1['count'] = counts
```

```python
df1.head(15)
```

```python
# get all unique subject_id as a list
subject_ids = df1['subject_id'].unique()
```

```python
len(subject_ids)
```

```python
keep = []
for subject_id in subject_ids:
    df2 = df1[df1['subject_id'] == subject_id]

    print(subject_id)
    print(df2.values)

    keeping = None
    for (subject_id, session_id, ckplus_labelled, count) in df2.values:
        if ckplus_labelled == 1:
            keeping = (subject_id, session_id)
            break

    if keeping is None:
        (subject_id, session_id, ckplus_labelled, count) = df2[df2['count'] == df2['count'].max()].values[0]
        keeping = (subject_id, session_id)

    print(keeping)
    keep.append(keeping)
    print()
```

```python
keep
```

```python
# from df keep only the rows that are in keep
ndf = df[df[['subject_id', 'session_id']].apply(tuple, axis=1).isin(keep)]
```

```python
ndf
```

```python
ndf.to_csv("disgust_filtered.csv", index=False)
```
