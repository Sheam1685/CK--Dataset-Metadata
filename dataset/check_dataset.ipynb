```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/sadness_filtered_top58.csv')
```

```python
# get all unique tuples of (subject_id, session_id, peak_emotion, ckplus_labelled)
df1 = df[['subject_id', 'session_id', 'peak_emotion', 'ckplus_labelled']].drop_duplicates().reset_index(drop=True)
```

```python
counts = (
    df.groupby(["subject_id", "session_id"])
    .size()
    .reset_index(name="count")["count"]
    .values.tolist()
)
df1['count'] = counts
```

```python
df1.info()
```

```python
# sort rows using custom order
df2 = df1.sort_values(by=['ckplus_labelled', 'count'], ascending=[False, False]).reset_index(drop=True)
```

```python
# only keep the first 50 rows
df2 = df2.head(58)
```

```python
pd.set_option("display.max_rows", None)
df2
```

```python
# only keep subject_id and session_id field
df3 = df2[['subject_id', 'session_id']]
```

```python
df3
```

```python
# only keep rows in df that are in df2
ndf = df[df[['subject_id', 'session_id']].apply(lambda x: tuple(x) in df3.values, axis=1)]
```

```python
ndf.to_csv('dataset/sadness_filtered_top58.csv', index=False)
```
