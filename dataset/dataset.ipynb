```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/ckplus_dataset.csv')
```

```python
df_ha = df[df['peak_emotion'] == 'happiness']
df_su = df[df['peak_emotion'] == 'surprise']
df_an = df[df['peak_emotion'] == 'anger']
df_sa = df[df['peak_emotion'] == 'sadness']
```

```python
ha_subjects = set(df_ha['subject_id'])
su_subjects = set(df_su['subject_id'])
an_subjects = set(df_an['subject_id'])
sa_subjects = set(df_sa['subject_id'])
```

```python
len(ha_subjects), len(su_subjects), len(an_subjects), len(sa_subjects)
```

```python
all_subjects = ha_subjects & su_subjects & an_subjects & sa_subjects
```

```python
len(all_subjects)
```

```python
df_ha = df_ha[df_ha['subject_id'].isin(all_subjects)]
df_su = df_su[df_su['subject_id'].isin(all_subjects)]
df_an = df_an[df_an['subject_id'].isin(all_subjects)]
df_sa = df_sa[df_sa['subject_id'].isin(all_subjects)]
```

```python
df_sa[['subject_id', 'session_id']].drop_duplicates()
```

```python
df_ha.to_csv('happiness.csv', index=False)
df_su.to_csv('surprise.csv', index=False)
df_an.to_csv('anger.csv', index=False)
df_sa.to_csv('sadness.csv', index=False)
```
